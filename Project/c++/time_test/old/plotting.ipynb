```python
%pylab inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = [10,100,1000,10000,100000, 1000000]
tree_av = np.array(x)

tree_av = [1.32751E-06,
1.30391E-06,
1.47939E-06,
1.81437E-06,
1.90592E-06, 
2.51055e-06]
tree_bal_av = [1.27983E-06,
1.03879E-06,
9.05991E-07,
1.13058E-06,
1.85037E-06, 1.81842e-06]

map_av = [8.52823E-07,
9.29594E-07,
8.93116E-07,
1.11723E-06,
1.8599E-06, 1.81985e-06]
tree_av = np.array(tree_av)*1e6
tree_bal_av = np.array(tree_bal_av)*1e6
map_av = np.array(map_av)*1e6
fit = numpy.polyfit(numpy.log(x)/log(2), tree_bal_av, 1)
```

```python
dev_tree = [4.87E-07,
3.65294E-07,
4.22983E-07,
2.16882E-07,
6.00678E-07, 5.74539e-07]
dev_tree_bal = [7.64E-07,
9.53674E-08,
4.18932E-07,
4.91368E-07,
3.91036E-07, 6.09339e-07]
dev_map = [4.77E-07,
9.53674E-08,
4.84377E-07,
5.21668E-07,
3.87384E-07,6.10292e-07]
dev_tree = np.array(dev_tree)*1e6
dev_tree_bal = np.array(dev_tree_bal)*1e6
dev_map = np.array(dev_map)*1e6
```

```python
semilogx(x,tree_av)
semilogx(x,tree_bal_av)
semilogx(x,map_av)
semilogx(x,fit[0]*log(x)/log(2)+fit[1])
ylabel('Time (ms) ')
xlabel('Size (N)')
title('Lookup time')

legend(['Tree unbalanced', 'Tree balanced', 'Map', 'Log2(x)'])
savefig('lookup_time.pdf')
```

```python
errorbar(x,tree_av, dev_tree)
errorbar(x,tree_bal_av, dev_tree_bal)
errorbar(x,map_av, dev_map)
errorbar(x, fit[0]*log(x)/log(2)+fit[1])
plt.xscale('log', nonposx='clip')
ylabel('Time (ms) ')
xlabel('Size (N)')
title('Lookup time')

legend(['Tree unbalanced', 'Tree balanced', 'Map', 'Log2(x)'])
savefig('lookup_time_we.pdf')
```
